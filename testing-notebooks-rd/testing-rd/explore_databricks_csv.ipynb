```python
import pandas as pd
import sqlite3

# Load CSV
df = pd.read_csv("ga4_events_2.csv")

# Connect to local SQLite file
conn = sqlite3.connect("local.db")

# Load DataFrame into SQLite table
df.to_sql("ga4_events", conn, if_exists="replace", index=False)
```

```python
df_ga4 = pd.read_sql("SELECT * FROM ga4_events LIMIT 5", conn)
```

```python
pd.read_sql("SELECT COUNT(*) AS total_events FROM ga4_events", conn)
```

```python
pd.read_sql("SELECT DISTINCT event_name FROM ga4_events", conn)
```

```python
pd.read_sql("""
SELECT event_name, COUNT(*) AS event_count
FROM ga4_events
GROUP BY event_name
ORDER BY event_count DESC
LIMIT 10
""", conn)
```
