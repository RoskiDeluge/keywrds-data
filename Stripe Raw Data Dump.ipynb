```python
%sql

-- 1) Create the schema (once)
CREATE SCHEMA IF NOT EXISTS stripe.stripe_data;

-- 2) Create an external table over all subscriptions files recursively
--    We use recursiveFileLookup + pathGlobFilter so you don’t need wildcards in LOCATION.
CREATE TABLE IF NOT EXISTS `stripe`.stripe_data.stripe_raw_data_v1
USING PARQUET
OPTIONS (
  path 'abfss://keywrds-stripe-data@keywrdsstripedata.dfs.core.windows.net/',
  recursiveFileLookup = true,
  pathGlobFilter = '*/subscriptions/livemode/*.parquet'
);
```

```python
display(
  spark.read.options(recursiveFileLookup="true", pathGlobFilter="*/livemode/subscriptions/*.parquet")
  .parquet("abfss://keywrds-stripe-data@keywrdsstripedata.dfs.core.windows.net/")
)
```

```python
df = spark.read.option("recursiveFileLookup", "true").parquet("abfss://keywrds-stripe-data@keywrdsstripedata.dfs.core.windows.net/2025*/livemode/subscriptions/")
df.printSchema()
```

```python
df.write.saveAsTable("stripe.stripe_data.tbl__stripe__initial_raw_subscriptions_data")
```
